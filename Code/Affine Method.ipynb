```python
def linear_congruence(a, b, m):
    if b == 0:
        return 0

    if a < 0:
        a = -a
        b = -b

    b %= m
    while a > m:
        a -= m

    return (m * linear_congruence(m, -b, a) + b) // a
```

```python
def kongruen_known(C,x,n):   
    m=-1
    f=True
    while f:
        if m<n:
            m=m+1
            if is_coprime(m, n):
                p=x*m % n
                if p==C:
                    f=False
        else:
            f=False
    return m

def known_kongruen(C1,P1,m,n):
    b=-1
    f=True
    while f:
        if b<n:
            b=b+1
            C=P1*m
            D=(C+b) % n
            if D==C1:
                f=False          
        else:
            f=False
    return b

def dekripsi_known(C,m,b,n):
    P=-1
    Plain=0
    f=True
    while f:
        if P<n:
            P=P+1
            Plain=(m*P+b) % n
            if Plain==C:
                f=False
        else:
            f=False
    return P
```
